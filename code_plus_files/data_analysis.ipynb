```python
import matplotlib.pyplot as plt 
import pandas as pd
import numpy as np
```

```python
def plot_data(csv_file, L_column, Q_column, T_column, gc, delta_epsilon, delta_sigma, x_min=None, x_max=None, L_subset=None, output_file="plot.pdf"):
    # Read data
    df = pd.read_csv(csv_file)

    # Calculate rescaled values
    df['t'] = (df[T_column] - gc) / gc  # For Z2 data
    a = 2 - delta_epsilon
    b = 2 * delta_sigma
    df['x'] = df['t'] * df[L_column] ** a
    df['y'] = df[Q_column] * df[L_column] ** b

    # Create figure
    fig, ax = plt.subplots(figsize=(10, 6))

    colors = ["red", "green", "blue", "orange", "purple", "brown", "pink", "gray", "olive", "cyan"]
    L_values = L_subset if L_subset else sorted(df[L_column].unique())

    for idx, L in enumerate(L_values):
        subset = df[df[L_column] == L]
        if x_min is not None and x_max is not None:
            subset = subset[(subset['x'] >= x_min) & (subset['x'] <= x_max)]
        color_idx = idx % len(colors)
        ax.scatter(subset['x'], subset['y'], marker='+', color=colors[color_idx], label=f'$L={L}$')

    # Axes labels
    ax.set_xlabel('Rescaled Coupling', fontsize=16)
    ax.set_ylabel('$L^{2\\beta/\\nu} \\langle Z_i Z_{i+L/2} \\rangle$', fontsize=16)

    # Grid and legend
    ax.grid(True, linestyle='--', alpha=0.7)
    ax.legend(fontsize=12)  # Removed "Legend" title

    # Title and layout
    plt.title('Scaling Collapse at Criticality', fontsize=18)
    plt.tight_layout()

    # Save to PDF
    plt.savefig(output_file, format="pdf")
    plt.show()

# Example Usage:
# plot_data('data.csv', 'L', 'Q', 'T', gc=0.5, delta_epsilon=0.3, delta_sigma=0.2, x_min=-0.8, x_max=0.8, output_file='plot.pdf')
```

```python
from scipy.interpolate import interp1d

def collapse_measure(params, df, L_column, Q_column, T_column, x_min, x_max, L_subset, power=1.0, debug=False):
    gc, delta_epsilon, delta_sigma = params
    df['t'] = (df[T_column] - gc) / gc  # Calculate reduced temperature 't'
    a = 2 - delta_epsilon
    b = 2 * delta_sigma

    df['rescaled_x'] = df['t'] * df[L_column] ** a  # Scale x by L^a
    df['rescaled_y'] = df[Q_column] * df[L_column] ** b  # Scale y by L^b

    all_residuals = []
    for L1 in L_subset:
        for L2 in L_subset:
            if L1 != L2:
                subset1 = df[df[L_column] == L1]
                subset2 = df[df[L_column] == L2]
                valid_data1 = subset1[(subset1['rescaled_x'] >= x_min) & (subset1['rescaled_x'] <= x_max)]
                valid_data2 = subset2[(subset2['rescaled_x'] >= x_min) & (subset2['rescaled_x'] <= x_max)]

                if not valid_data1.empty and not valid_data2.empty:
                    # Handle duplicates by averaging over y
                    valid_data2 = valid_data2.groupby('rescaled_x')['rescaled_y'].mean().reset_index()

                    # Use linear interpolation if there are not enough points for cubic
                    kind = 'cubic' if len(valid_data2) > 3 else 'linear'
                    interp_func = interp1d(valid_data2['rescaled_x'], valid_data2['rescaled_y'], kind=kind,
                                           fill_value='extrapolate', bounds_error=False)

                    if debug:
                        print(f"Debugging info for L2 = {L2}: Number of points = {len(valid_data2)}, Interpolation kind = {kind}")

                    for xi, yi in zip(valid_data1['rescaled_x'], valid_data1['rescaled_y']):
                        if x_min <= xi <= x_max:
                            interpolated_yi = interp_func(xi)
                            residual = np.abs(yi - interpolated_yi) ** power
                            all_residuals.append(residual)

    normalized_residual = np.mean(all_residuals) ** (1 / power)

    return normalized_residual
```

```python
from scipy.optimize import least_squares

def compute_dimensions(args):
    df, L_column, Q_column, T_column, x_min, x_max, L_subset, power, initial_guess, bounds = args
    result = least_squares(
        collapse_measure,
        initial_guess,
        args=(df, L_column, Q_column, T_column, x_min, x_max, L_subset, power),
        bounds=bounds
    )
    return result.x

# we can perform statistical bootstrapping of our original dataset to obtain uncertainities 

# note that since we are sampling with replacement it is essential to pick x_min and x_max 
# such that even for the largest length in L_subset there are enough points for iterpolation
# given the data we have the lowest suggest width is [-0.5,0.5], but [-1,1] works even better
# to investigate this in detail turn the debug flag on in collapse_measure function  

def bootstrap_iteration(args, debug=False):
    df, L_column, Q_column, T_column, x_min, x_max, L_subset, power, initial_guess, bounds = args
    sampled_df = df.sample(frac=1.0, replace=True, random_state=np.random.RandomState())
    if debug:
        print(f"Sampled indices: {sampled_df.index.tolist()[:5]}")  # Print first 5 indices of the sampled data # this is to verify that there is enough randomness in bootstrap sampling  
        #print(f"Number of unique indices: {sampled_df.index.nunique()}")
        #print(f"Total number of indices: {len(sampled_df.index)}")
    args = sampled_df, L_column, Q_column, T_column, x_min, x_max, L_subset, power, initial_guess, bounds
    return compute_dimensions(args)
```

```python
from joblib import Parallel, delayed  # we use joblib instead of built-in python parallel threading since it is better at handling random seeds in parallel 

def bootstrap_params(df, L_column, Q_column, T_column, initial_guess, bounds, x_min, x_max, L_subset, power, n_iterations=10): # 100 bootstrap iterations is sufficient to estimate the standard deviations 
    args = [(df, L_column, Q_column, T_column, x_min, x_max, L_subset, power, initial_guess, bounds) for _ in range(n_iterations)]
    
    # Use Joblib Parallel to manage independent processes
    results = Parallel(n_jobs=-1)(delayed(bootstrap_iteration)(arg, debug=False) for arg in args)
    param_samples = np.array(results)

    mean_params = np.mean(param_samples, axis=0)
    std_params = np.std(param_samples, axis=0)

    mean_params = np.round(mean_params, 5)
    std_params = np.round(std_params, 5)

    print("gc, delta_epsilon, delta_sigma:", mean_params)
    print("err_gc, err_epsilon, err_sigma:", std_params)
    return mean_params, std_params
```

```python
initial_guess = [1.00, 1.00, 0.125]  # Example initial guesses for gc, delta_epsilon, delta_sigma
bounds = ([0.98, 0.98, 0.11], [1.02, 1.02, 0.13])  # Bounds for the parameters
df = pd.read_csv(file_path)

# this takes a while to run since there are many optimizations being performed 
mean_params, std_params = bootstrap_params(df, 'L', 'Mean_Two_Point_Correlation', 'g', initial_guess, bounds, x_min=-0.5, x_max=0.5, L_subset=[12,16,20], power=1.0, n_iterations=200)
```

```python
import os
import csv
import json

def generate_summary_csv(base_dir: str, output_csv: str):
    """
    Walks through the base directory, collects data from JSON summaries, and writes a consolidated CSV.

    Args:
        base_dir (str): Directory containing results with subdirectories for each L and g.
        output_csv (str): Path to the consolidated summary CSV.
    """
    # Define CSV headers
    headers = [
        "L", "g", "E_exact", "E_vmc_real", "Var_E",
        "Mean_Two_Point_Correlation", "Var_Two_Point_Correlation",
        "Exact_Two_Point_Correlation", "Exact_Two_Point_Correlation_Variance", "rel_error"
    ]
    
    # Open the output CSV file
    with open(output_csv, "w", newline="") as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(headers)
        
        # Walk through the base directory
        for root, _, files in os.walk(base_dir):
            for file in files:
                if file.endswith("summary.json"):
                    summary_path = os.path.join(root, file)
                    
                    # Read the JSON file
                    with open(summary_path, "r") as f:
                        try:
                            summary = json.load(f)
                            
                            # Extract data
                            L = summary.get("L", "N/A")
                            g = summary.get("g", "N/A")
                            g = round(g, 4)
                            E_exact = summary.get("E_exact", "N/A")
                            E_vmc_real = summary.get("E_vmc_real", "N/A")
                            Var_E = summary.get("Var_E", "N/A")
                            Mean_Two_Point_Correlation = summary.get("Mean_Two_Point_Correlation", "N/A")
                            Var_Two_Point_Correlation = summary.get("Var_Two_Point_Correlation", "N/A")
                            Exact_Two_Point_Correlation = summary.get("Exact_Two_Point_Correlation", "N/A")
                            Exact_Two_Point_Correlation_Variance = summary.get("Exact_Two_Point_Correlation_Variance", "N/A")
                            rel_error = summary.get("rel_error", "N/A")
                            
                            # Write data to CSV
                            writer.writerow([
                                L, g, E_exact, E_vmc_real, Var_E,
                                Mean_Two_Point_Correlation, Var_Two_Point_Correlation,
                                Exact_Two_Point_Correlation, Exact_Two_Point_Correlation_Variance, rel_error
                            ])
                        except json.JSONDecodeError as e:
                            print(f"Error reading JSON file {summary_path}: {e}")


# Define base directory containing results
base_dir = "results"

# Define output CSV file
output_csv = "summary.csv"

# Generate the summary CSV
generate_summary_csv(base_dir, output_csv)
print(f"Summary CSV generated: {output_csv}")
```

```python
import os
import json
import matplotlib.pyplot as plt
from cycler import cycler

def find_logs_for_g(base_dir, target_g):
    """
    Walks through the base directory and finds all state.log files for the specified g value.
    """
    log_files = {}
    for root, dirs, files in os.walk(base_dir):
        for file in files:
            if file == "state.log":
                # Extract lattice size and g from the directory structure
                path_parts = root.split(os.sep)
                L = None
                g_value = None
                for part in path_parts:
                    if part.startswith("L_"):
                        L = int(part.split("_")[1])
                    if part.startswith("g_"):
                        g_value = float(part.split("_")[1])
                # Check if this log corresponds to the target g
                if L is not None and g_value is not None and abs(g_value - target_g) < 1e-4:
                    log_files[L] = os.path.join(root, file)
    return log_files

import os
import json
import matplotlib.pyplot as plt
from cycler import cycler

def plot_convergence(log_files, target_g, output_file="convergence_plot.pdf"):
    """
    Reads log files and plots convergence of E/L vs steps for the specified g.
    
    Args:
        log_files (dict): Dictionary with lattice sizes as keys and log file paths as values.
        target_g (float): The target coupling constant g for labeling.
        output_file (str): Path to save the output PDF file.
    """
    # Set plot style
    plt.rcParams.update({
        "font.size": 14,
        "axes.labelsize": 16,
        "axes.titlesize": 18,
        "legend.fontsize": 14,
        "grid.color": "gray",
        "grid.linestyle": "--",
        "grid.linewidth": 0.6,
        "axes.prop_cycle": cycler("color", ["red", "blue", "green", "orange", "purple"]),
    })

    fig, ax = plt.subplots(figsize=(8, 6))

    for L, log_file in sorted(log_files.items()):
        # Read log data
        with open(log_file, "r") as file:
            log_data = json.load(file)

        # Extract iteration and energy data
        iterations = log_data["Energy"]["iters"]
        energy_mean_real = log_data["Energy"]["Mean"]["real"]

        # Normalize energy by lattice size
        energy_per_L = [E / L for E in energy_mean_real]

        # Plot energy per site
        ax.plot(iterations, energy_per_L, label=f"$L={L}$", linewidth=2)

    # Set axis labels and title
    ax.set_xlabel("Iteration Steps")
    ax.set_ylabel(r"$E/L$")
    ax.set_title("Energy per Site vs Iteration Steps at $g=2.0$")

    # Add legend and grid
    ax.legend(loc="upper right", frameon=True)  # Boxed legend
    ax.grid(True, alpha=0.6)

    # Tight layout for cleaner appearance
    plt.tight_layout()

    # Save the plot to a PDF
    plt.savefig(output_file, format="pdf")
    plt.show()

# Example Usage
base_dir = "results"  # Replace with the correct path
target_g = 2.00  # Target coupling constant

# Find logs and plot convergence
log_files = find_logs_for_g(base_dir, target_g)
if log_files:
    plot_convergence(log_files, target_g, output_file="convergence.pdf")
else:
    print(f"No logs found for g = {target_g:.2f}")
```
